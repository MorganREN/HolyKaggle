```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = pd.read_csv('./train.csv', index_col='Id')
X_test_full = pd.read_csv('./test.csv', index_col='Id')

X.head()
```

```python
X.dropna(axis=0, subset=['SalePrice'], inplace=True)
y = X['SalePrice']
X.drop(['SalePrice'], axis=1, inplace=True)

print(X.shape)

X.head()
```

```python
print(y.shape)

y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train_full, X_valid_full, y_train_full, y_valid_full = train_test_split(X, y, train_size=0.8, test_size=0.2)

print(X_train_full.shape)
print(X_valid_full.shape)
print(y_train_full.shape)
print(y_valid_full.shape)
```

```python
low_cal_col = [cname for cname in X_train_full.columns if X_train_full[cname].nunique() < 10 and 
               X_train_full[cname].dtype == 'object']

low_cal_col  # columns which can be one-hot coded
```

```python
numerical_col = [cname for cname in X_train_full.columns if X_train_full[cname].dtype in ['int64', 'float64']]

numerical_col
```

```python
all_cols = low_cal_col + numerical_col

X_train = X_train_full[all_cols].copy()
X_valid = X_valid_full[all_cols].copy()
X_test = X_test_full[all_cols].copy()
```

```python
# Pipeline for preprocessing
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

numerical_transformer = SimpleImputer(strategy='mean')

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('one-hot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_col),
        ('cat', categorical_transformer, low_cal_col)
    ]
)
```

```python
X_train.head()
```

```python
# Build the model
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error

model = XGBRegressor(n_estimators=1000, learning_rate=0.01)

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', model)
])

pipeline.fit(X_train, y_train_full)

pred = pipeline.predict(X_valid)

mae = mean_absolute_error(y_valid_full, pred)

print("The MAE is:", mae)
```

## Use cross Validation

```python
# Use cross validation
from sklearn.model_selection import cross_val_score

# Possible values for n_estimator
n_estimators_list = [100*i for i in range(10)]

for n in n_estimators_list:
    model_CV = XGBRegressor(n_estimators=n, learning_rate=0.01)
    
    pipeline_CV = Pipeline(steps=[
                        ('preprocessor', preprocessor),
                        ('model', model_CV)
                    ])

    
    scores = -1 * cross_val_score(pipeline_CV, X[all_cols], y,
                                 cv=5,
                                 scoring='neg_mean_absolute_error')
    
    print("Mean MAE score for n_estimator = {} is {}".format(n, scores.mean()))
```
